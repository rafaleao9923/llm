```python
import random

# Lists of product attributes
product_types = ['Smartphone', 'Laptop', 'Headphones', 'Smartwatch', 'Tablet']
brands = ['TechPro', 'SmartGear', 'ElectroMax', 'FutureTech', 'InnovateCorp']
colors = ['Black', 'Silver', 'White', 'Blue', 'Red']
materials = ['Aluminum', 'Plastic', 'Glass', 'Titanium', 'Carbon Fiber']

# Feature templates
smartphone_features = [
    "Display: {size} HD AMOLED screen",
    "Camera: {mp}MP main camera with Night Mode",
    "Battery: {capacity}mAh with fast charging",
    "Storage: {storage}GB internal storage",
    "Processor: {processor} chipset"
]

laptop_features = [
    "{size} Full HD display",
    "{processor} processor",
    "{ram}GB RAM",
    "{storage}GB SSD storage",
    "Up to {battery_life} hours of battery life"
]

headphones_features = [
    "Active Noise Cancellation",
    "{playtime} hours of playtime",
    "Bluetooth {bluetooth_version}",
    "Water resistance: IPX{rating}",
    "{driver_size}mm dynamic drivers"
]

smartwatch_features = [
    "{size} OLED display",
    "Heart rate and {sensor} sensor",
    "Battery life: up to {battery_life} days",
    "Water resistance: {water_rating}",
    "GPS and {connectivity} connectivity"
]

tablet_features = [
    "{size} Retina display",
    "{processor} chip",
    "{storage}GB storage",
    "{camera}MP rear camera, {front_camera}MP front camera",
    "Battery life: up to {battery_life} hours"
]

def generate_description(product_type):
    brand = random.choice(brands)
    color = random.choice(colors)
    material = random.choice(materials)

    description = f"Product: {brand} {product_type}\n"
    description += f"Color: {color}\n"
    description += f"Material: {material}\n"
    description += "Features:\n"

    if product_type == 'Smartphone':
        for feature in smartphone_features:
            if "Display" in feature:
                description += feature.format(size=random.choice(['5.8"', '6.1"', '6.5"'])) + "\n"
            elif "Camera" in feature:
                description += feature.format(mp=random.choice([12, 16, 32, 48, 64])) + "\n"
            elif "Battery" in feature:
                description += feature.format(capacity=random.choice([3000, 3500, 4000, 4500])) + "\n"
            elif "Storage" in feature:
                description += feature.format(storage=random.choice([64, 128, 256, 512])) + "\n"
            elif "Processor" in feature:
                description += feature.format(processor=random.choice(['A14 Bionic', 'Snapdragon 888', 'Exynos 2100'])) + "\n"

    elif product_type == 'Laptop':
        for feature in laptop_features:
            if '"' in feature:
                description += feature.format(size=random.choice(['13.3"', '14"', '15.6"', '17"'])) + "\n"
            elif "processor" in feature:
                description += feature.format(processor=random.choice(['Intel Core i5', 'Intel Core i7', 'AMD Ryzen 5', 'AMD Ryzen 7'])) + "\n"
            elif "RAM" in feature:
                description += feature.format(ram=random.choice([8, 16, 32])) + "\n"
            elif "storage" in feature:
                description += feature.format(storage=random.choice([256, 512, 1024])) + "\n"
            elif "battery life" in feature:
                description += feature.format(battery_life=random.choice([8, 10, 12, 14])) + "\n"

    elif product_type == 'Headphones':
        for feature in headphones_features:
            if "playtime" in feature:
                description += feature.format(playtime=random.choice([20, 24, 30, 36])) + "\n"
            elif "Bluetooth" in feature:
                description += feature.format(bluetooth_version=random.choice(['5.0', '5.1', '5.2'])) + "\n"
            elif "IPX" in feature:
                description += feature.format(rating=random.choice([4, 5, 7])) + "\n"
            elif "drivers" in feature:
                description += feature.format(driver_size=random.choice([40, 43, 45, 50])) + "\n"
            else:
                description += feature + "\n"

    elif product_type == 'Smartwatch':
        for feature in smartwatch_features:
            if '"' in feature:
                description += feature.format(size=random.choice(['1.2"', '1.3"', '1.4"'])) + "\n"
            elif "sensor" in feature:
                description += feature.format(sensor=random.choice(['SpO2', 'ECG', 'blood pressure'])) + "\n"
            elif "Battery life" in feature:
                description += feature.format(battery_life=random.choice([2, 3, 5, 7])) + "\n"
            elif "Water resistance" in feature:
                description += feature.format(water_rating=random.choice(['5 ATM', '10 ATM', 'IP68'])) + "\n"
            elif "connectivity" in feature:
                description += feature.format(connectivity=random.choice(['LTE', 'Wi-Fi', 'NFC'])) + "\n"

    elif product_type == 'Tablet':
        for feature in tablet_features:
            if '"' in feature:
                description += feature.format(size=random.choice(['8"', '10.2"', '11"', '12.9"'])) + "\n"
            elif "chip" in feature:
                description += feature.format(processor=random.choice(['A12 Bionic', 'A14 Bionic', 'M1'])) + "\n"
            elif "storage" in feature:
                description += feature.format(storage=random.choice([32, 64, 128, 256])) + "\n"
            elif "camera" in feature:
                description += feature.format(camera=random.choice([8, 12, 16]), front_camera=random.choice([5, 7, 12])) + "\n"
            elif "Battery life" in feature:
                description += feature.format(battery_life=random.choice([10, 12, 14])) + "\n"

    return description

# Generate dataset
num_descriptions = 1000
product_descriptions = [generate_description(random.choice(product_types)) for _ in range(num_descriptions)]

print(f"Generated {num_descriptions} product descriptions")
print("\nSample product description:")
print(product_descriptions[0])
```

```python
import torch
from transformers import GPT2LMHeadModel, GPT2Tokenizer, DataCollatorForLanguageModeling
from transformers import Trainer, TrainingArguments
from torch.utils.data import Dataset

# Check if CUDA is available
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

# Load pre-trained model and tokenizer
model_name = "gpt2"
model = GPT2LMHeadModel.from_pretrained(model_name).to(device)
tokenizer = GPT2Tokenizer.from_pretrained(model_name)
tokenizer.pad_token = tokenizer.eos_token

# Custom dataset class
class ProductDescriptionDataset(Dataset):
    def __init__(self, texts, tokenizer, max_length):
        self.encodings = tokenizer(texts, truncation=True, padding=True, max_length=max_length, return_tensors="pt")

    def __getitem__(self, idx):
        item = {key: val[idx].clone().detach() for key, val in self.encodings.items()}
        item['labels'] = item['input_ids'].clone()
        return item

    def __len__(self):
        return len(self.encodings.input_ids)

# Prepare your dataset
train_dataset = ProductDescriptionDataset(product_descriptions, tokenizer, max_length=128)

# Data collator
data_collator = DataCollatorForLanguageModeling(
    tokenizer=tokenizer, mlm=False,
)

# Training arguments
training_args = TrainingArguments(
    output_dir="./fine_tuned_model",
    overwrite_output_dir=True,
    num_train_epochs=3,
    per_device_train_batch_size=4,
    save_steps=10_000,
    save_total_limit=2,
    fp16=True,  # Enable mixed precision training
    no_cuda=False,  # Ensure GPU is used
)

# Initialize Trainer
trainer = Trainer(
    model=model,
    args=training_args,
    data_collator=data_collator,
    train_dataset=train_dataset,
)

# Fine-tune the model
trainer.train()

# Save the fine-tuned model
model.save_pretrained("./fine_tuned_model")
tokenizer.save_pretrained("./fine_tuned_model")

# Generate a product description
def generate_description(prompt, model, tokenizer, max_new_tokens=100):
    # Encode the prompt
    inputs = tokenizer.encode_plus(
        prompt,
        add_special_tokens=True,
        return_tensors="pt",
        padding=True,
        truncation=True
    )
    input_ids = inputs["input_ids"].to(device)
    attention_mask = inputs["attention_mask"].to(device)

    # Generate output
    output = model.generate(
        input_ids=input_ids,
        attention_mask=attention_mask,
        max_new_tokens=max_new_tokens,
        num_return_sequences=1,
        no_repeat_ngram_size=2,
        top_k=50,
        top_p=0.95,
        temperature=0.7,
        do_sample=True,
        pad_token_id=tokenizer.eos_token_id
    )

    # Decode and return the generated text
    return tokenizer.decode(output[0], skip_special_tokens=True)

# Example usage
fine_tuned_model = GPT2LMHeadModel.from_pretrained("./fine_tuned_model").to(device)
fine_tuned_tokenizer = GPT2Tokenizer.from_pretrained("./fine_tuned_model")

prompt = "Product: TechPro Smartwatch\nFeatures:"
generated_description = generate_description(prompt, fine_tuned_model, fine_tuned_tokenizer)
print("Generated Product Description:")
print(generated_description)

# Clean up
torch.cuda.empty_cache()
```
